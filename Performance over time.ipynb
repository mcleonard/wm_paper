```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import cPickle as pkl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import memory as my 
from spikesort import catalog
```

Let's select and open the behavior data.

```python
database = '//Users/mat/Dropbox/Data/LLRR.sqlite'
log = catalog.open(database)
sessions = log.query(log.Session).\
           filter(log.Session.rat == 'MAT21B').\
           order_by(log.Session.date).\
           all()
            
bhv_files = (my.utils.filepath_from_dir(each.path, 'bhv') for each in sessions)
bhv_data = (pkl.load(open(each[0])) for each in bhv_files)
dfs = [my.bhv.build_data(each) for each in bhv_data]
```

The data for each session is stored in a Pandas DataFrame.

```python
dfs[0].head()
```

For each session, we want to calculate the performance (percent correct trials), for both cued and uncued trials. We also only want only the trials that took less than 20 seconds to initiate a new trial.

```python
results = []
for df in dfs:
    df['delay'] = df['onset'] - df['PG out']
    df = df[df['delay'] <= 20]
    perfs = df.groupby('block')['hits'].agg({'performance':np.mean})
    results.append(perfs*100)
results = pd.concat(results, axis=1)
```

```python
fig, ax = plt.subplots(figsize=(5,3))

# Plot the performance
x = range(1,results.shape[1]+1)
ax.plot(x, results.ix[my.RM], '-o', color='grey', 
        ms=7, mfc='w', mew=2, mec='grey', label='Cued')
ax.plot(x, results.ix[my.WM], '-o', color='k', 
        ms=7, mfc='w', mew=2, label='Uncued')

# Labels and things
leg = ax.legend(loc=(0.63, 0.20))
ax.set_ylim(40,105)
ax.plot(ax.get_xlim(), [50, 50], '--k', alpha=0.5)
ax.set_xlabel('Sessions after recovery')
ax.set_ylabel('Percent correct trials')
ax = my.utils.pub_format(ax, legend=leg, yticks=7)
```
