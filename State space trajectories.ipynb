```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from collections import defaultdict
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import memory as my
from spikesort import catalog
```

```python
database = '//Users/mat/Dropbox/Data/LLRR.sqlite'
log = catalog.open(database)
units = log.query(log.Unit).join(log.Session).\
        filter(log.Unit.cluster != 0).\
        filter(log.Unit.rate > 0.5).\
        filter(log.Session.rat != 'ALY2B').\
        all()
# Unfortunately, timelocking the data takes a long time.
# I found that loading the data into memory is what times the most time.
lkr = my.timelock.timelock(units)
lkr.lock('onset')
```

First, for the PCA part, let's not care about goals and such yet. Let's just get the PCA components.

```python
lkr.lock('C in')
states = pd.DataFrame()
for i, unit in enumerate(units):
    udata = lkr[unit.id].groupby(['PG outcome', 'FG outcome']).get_group((my.HIT, my.HIT))
    spikes = my.analyze.time_histogram(udata, limit=(-1, 1), bin_width=0.05)*0.05
    states = pd.concat([states, spikes.mean(axis=0)], axis=1)
states.index = spikes.columns
states.columns = range(len(units))
```

```python
from sklearn.decomposition import PCA
```

```python
plt.plot(states[10], states[2], '-k', zorder=1)
plt.scatter(states[10], states[2], s=40, c=plt.cm.RdBu(np.linspace(0,1,40)), zorder=2, edgecolor='k')
```

```python
# De-mean
states = (states.T - states.mean(axis=1)).T
```

```python
pca = PCA(n_components=6)
pca.fit(states)
pca.explained_variance_ratio_.sum()
reduced = pca.transform(states)
pcs = pca.components_
```

```python
plt.plot(pca.explained_variance_ratio_, '-ok')
```

```python
cs = (0,2)
plt.plot(reduced[:,cs[0]], reduced[:,cs[1]], '-k', zorder=1)
plt.scatter(reduced[:,cs[0]], reduced[:,cs[1]], s=40, c=plt.cm.RdBu(np.linspace(0,1,40)), zorder=2, edgecolor='k')
```

Okay, now I have my data in a 6 dimensional reduced space. Now, I have to do some linear regression and get past goal, future goal, and switch axes. Then we can use those as the targets for the regression.
