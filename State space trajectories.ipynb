```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from collections import defaultdict
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import memory as my
from spikesort import catalog
```

```python
database = '//Users/mat/Dropbox/Data/LLRR.sqlite'
log = catalog.open(database)
units = log.query(log.Unit).join(log.Session).\
            filter(log.Unit.cluster != 0).\
            filter(log.Session.rat == 'MAT21B').\
            filter(log.Unit.rate >= 0.5).all()
# Unfortunately, timelocking the data takes a long time.
# I found that loading the data into memory is what times the most time.
lkr = my.timelock.timelock(units)
lkr.lock('onset')
```

First, for the PCA part, let's not care about goals and such yet. Let's just get the PCA components.

```python
lkr.lock('C out')
states = pd.DataFrame()
for i, unit in enumerate(units):
    udata = lkr[unit.id].groupby(['PG outcome', 'FG outcome']).get_group((my.HIT, my.HIT))
    spikes = my.analyze.time_histogram(udata, limit=(-1, 0.6), bin_width=0.05)*0.05
    states = pd.concat([states, spikes.mean(axis=0)], axis=1)
states.index = spikes.columns
states.columns = range(len(units))

# De-mean
states = (states.T - states.mean(axis=1)).T
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=6)
pca.fit(states)
pca.explained_variance_ratio_.sum()
reduced = pca.transform(states)
pcs = pca.components_
```

```python
plt.plot(pca.explained_variance_ratio_, '-ok')
```

```python
cs = (1,2)
plt.plot(reduced[:,cs[0]], reduced[:,cs[1]], '-k', zorder=1)
plt.scatter(reduced[:,cs[0]], reduced[:,cs[1]], s=40, c=plt.cm.RdBu(np.linspace(0,1,40)), zorder=2, edgecolor='k')
plt.xlim(-1,1)
plt.ylim(-1,1)
```

Okay, now I have my data in a 6 dimensional reduced space. Now, I have to do some linear regression and get past goal, future goal, and switch axes. First, in one time bin, for each neuron find the average rate for PG = left, PG = right, and so on. So we'll get a population state for PG = left and a population state for PG = right. Then we can use those as the targets for the regression. I'm going to calculate the PG axis at center in, then the switch axis at center out, then the FG ("choice") axis at FG in.

```python
from sklearn.linear_model import LinearRegression
axes = {}
for time_point, var in zip(['C in', 'C out', 'FG in'], ['PG port', 'switch', 'FG port']):
    lkr.lock(time_point)
    variable = var
    
    # Construct states grouped by the task-related variable
    results = pd.DataFrame(np.array([1,2]), columns=[variable])
    for i, each in enumerate(units):
        udata = lkr[each.id].groupby(['PG outcome', 'FG outcome']).get_group((my.HIT, my.HIT))
        udata['spikes'] = my.time_histogram(udata, limit=(-0.025, 0.025), bin_width=0.05)*0.05
        udata['switch'] = ~(udata['PG port'] == udata['FG port']) + 1
        df = udata.groupby([variable])['spikes'].agg({i:np.mean})
        df = df.reset_index()
        results = results.merge(df)
        
    # Do the linear regression and save the coefficient vector    
    y = pca.transform(results.icol(range(1,results.shape[1])))
    X = 2*results[[variable]].values - 3 # Shift from 1,2 to -1,1
    lrm = LinearRegression()
    lrm.fit(X, y)
    axes[var] = lrm.coef_
```

Okay, now, get states for each PG, FG pair, then project them onto the task-related axes.

```python
lkr.lock('C out')
states = dict(zip([(1,1), (1,2), (2,1), (2,2)],
                  [pd.DataFrame() for _ in range(4)]))
for i, unit in enumerate(units):
    udata = lkr[unit.id].groupby(['PG outcome', 'FG outcome']).get_group((my.HIT, my.HIT))
    for group, gdata in udata.groupby(['PG port', 'FG port']):
        spikes = my.analyze.time_histogram(gdata, limit=(-1, 0.6), bin_width=0.05)*0.05
        states[group] = pd.concat([states[group], spikes.mean(axis=0)], axis=1)

for each in states:
    states[each].index = spikes.columns
    states[each].columns = range(len(units))
```

```python
y = pca.transform(states[(2,2)]).dot(axes['PG port'])
x = pca.transform(states[(2,2)]).dot(axes['FG port'])
plt.plot(x, y, '-k', zorder=1)
plt.scatter(x, y, s=40, c=plt.cm.RdBu(np.linspace(0,1,40)), zorder=2, edgecolor='k')
#plt.xlim(-0.2, 0.2)
#plt.ylim(-0.6, 0.6)
```

```python
def color_time_points(c):
    c[states[(1,1)].index==-0.35] = (1,0,1,1)
    c[states[(1,1)].index==0.0] = (0,1,0,1)
    #c[states[(1,1)].index==0.5] = (0,0,0,1)
    return c
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6,6), sharex=True, sharey=True)
goal_dict = {my.LEFT:'Left', my.RIGHT:'Right'}
for each, ax in zip(states, axs.ravel()):
    y = pca.transform(states[each]).dot(axes['PG port'])
    x = pca.transform(states[each]).dot(axes['FG port'])
    
    ax.plot(x, y, '-k', zorder=2, alpha=0.5)
    ax.scatter(x, y, s=40, zorder=3, edgecolor='k',
                c=color_time_points(plt.cm.RdBu(np.linspace(0,1,len(x)))))
    ax.set_xlim(-1.5, 1)
    ax.set_ylim(-1, 1)
    ax.plot(ax.get_xlim(), [0,0], '--k', alpha=0.3, zorder=1)
    ax.plot([0,0], ax.get_ylim(), '--k', alpha=0.3, zorder=1)
    ax = my.utils.pub_format(ax, xticks=3)
    ax.text(ax.get_xlim()[0]*0.9,ax.get_ylim()[1]*0.9,'{} to {}'.format(goal_dict[each[0]], goal_dict[each[1]]))
axs[1,0].set_xlabel("Choice")
axs[1,1].set_xlabel("Choice")
axs[0,0].set_ylabel("Past goal")
axs[1,0].set_ylabel("Past goal")
fig.tight_layout(pad=0.1, h_pad=1, w_pad=1)
```
