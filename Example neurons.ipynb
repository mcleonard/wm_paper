```python
%matplotlib inline

# This increases resolution of inline images
%config InlineBackend.figure_format = 'retina' 

#import warnings
#warnings.filterwarnings("ignore", category=DeprecationWarning) 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp

import memory as my
from spikesort import catalog

# Need to do this because %matplotlib inline sets the font size to 10,
# ignoring the rcParam file
plt.rcParams['font.size'] = 12
```

I have the metadata about the units I've recorded stored in an SQLite database. I can select only the units I want to analyze. In the sorting procedure, cluster 0 is always a noise cluster, so I am filtering out all of those. I also want to analyze only units with average firing rates greater than 0.5 Hz. Once the units are selected, I can express all the spikes times relative to some behvioral event (typically called timelocking). In this case, I locked the spike times to the go sound onset. We can see that there is data from 30 different recording sessions.

```python
database = '//Users/mat/Dropbox/Data/LLRR.sqlite'
log = catalog.open(database)
units = log.query(log.Unit).join(log.Session).\
        filter(log.Unit.cluster != 0).\
        filter(log.Unit.rate > 0.5).\
        filter(log.Session.rat != 'ALY2B').\
        all()
print("{} units selected".format(len(units)))
lkr = my.timelock.timelock(units)
lkr.lock('onset')
```

I want to make a figure showing the spike times on different trials, this type of figure is typically called a raster plot. I'm going to split the figure into four small plots, one for each trajectory (left to right, left to left, etc). 

The function `my.analyze.time_histogram` bins each trial between times `limit` with bin width equal to `bin_width`. It returns a Pandas data frame where each row is one trial, and each column is one time bin, and the values are the average rate on the trial in the bin.

```python
lkr.lock('C out') # Center out event
unit = 236
limits = (-1, 1.5)
port_map = {my.LEFT:'Left', my.RIGHT:'Right'}
fig = plt.figure(num=1, figsize=(10,4))

# Select trials where the rat made the correct response twice in a row
data = lkr[unit].groupby(('PG outcome', 'FG outcome')).get_group((my.HIT, my.HIT))
trajectories = data.groupby(('PG port', 'FG port'))
for ii, ((pg, fg), tdata) in enumerate(trajectories):
    
    # Sort the trials first, for visual clarity
    tdata = tdata.sort(columns='FG in')
    
    ########################
    # Make the histogram plot
    ax_bar = plt.subplot2grid((3,4), (0,ii))
    rates = my.analyze.time_histogram(tdata, limit=limits, bin_width=0.05)
    ax_bar.bar(rates.columns, rates.mean(), width=0.05, facecolor='w', edgecolor='k', zorder=1)
    ax_bar.set_xticklabels('')
    ax_bar.set_ylim(0,15)
    ax_bar.plot([0,0], ax_bar.get_ylim(), '--k', alpha=0.5, zorder=3)
    ax_bar.set_title('{} to {}'.format(port_map[pg], port_map[fg]), size='medium')
    
    # Adding the colored bars here, they indicate the middle 50% of the event times.
    # Here the events are entering the center port (C in) and entering the goal port (FG in).
    lq, uq = tdata['C in'].quantile(q=[0.25, .75])
    ax_bar.fill_betweenx(ax_bar.get_ylim(), lq, uq, color='steelblue', alpha=0.25, zorder=3)
    lq, uq = tdata['FG in'].quantile(q=[0.25, .75])
    ax_bar.fill_betweenx(ax_bar.get_ylim(), lq, uq, color='orange', alpha=0.25, zorder=3)
    
    ax_bar.set_xlim(*limits)
    ax_bar = my.utils.pub_format(ax_bar, yticks=3, xticks=2)
    
    ######################
    # Make the raster plot
    timestamps = np.hstack(tdata['timestamps'].values)
    trial_nums = np.hstack([[jj]*len(row) for jj, row in enumerate(tdata['timestamps'])])
    ax_scatter = plt.subplot2grid((3,4), (1,ii), rowspan=2)
    ax_scatter.scatter(timestamps, trial_nums, marker='|', color='k', s=5)
    ax_scatter.scatter(tdata['C in'].values, range(len(tdata)), marker='.', c='steelblue', edgecolors='steelblue')
    ax_scatter.scatter(tdata['FG in'].values, range(len(tdata)), marker='.', c='orange', edgecolors='orange')
    
    ax_scatter.set_xlim(*limits)
    ax_scatter.set_ylim(0, 100)
    ax_scatter.plot([0,0], ax_scatter.get_ylim(), '--k', alpha=0.5)
    ax_scatter = my.utils.pub_format(ax_scatter, yticks=3, xticks=3)
    
    # Turn off yticks and labels for all but the leftmost axes
    if ii > 0:
        ax_bar.set_yticklabels('')
        ax_scatter.set_yticklabels('')
    else:
        ax_bar.set_ylabel('Activity \n(spikes/s)')
        ax_scatter.set_ylabel('Trial')
    
    if ii == 1:
        ax_scatter.set_xlabel('Time from response (seconds)', horizontalalignment='left')
    
fig.tight_layout(pad=0.1, w_pad=-12, h_pad=1)
```

```python
def rate_estimate_bayes(t, n, alpha=0.5, beta=0):
    """ Returns the mean and 95% credible region for a model with a
        Poisson likelihood and the Jeffreys prior, ~sqrt(1/lambda).
        As such, the posterior is Gamma(0.5 + t, n), where t is the total 
        number of events and n is the number of trials.
    """
    posterior = sp.stats.gamma(alpha + t, scale=(beta + n)**-1)
    return posterior.mean(), posterior.interval(0.95)
```

```python
lkr.lock('C out')
examples = [53, 57, 236, 313]
fig, axes = plt.subplots(nrows=len(examples), sharex=True, figsize=(5,6))
bin_width = 0.1
labels = {my.LEFT:'Left', my.RIGHT:'Right'}
for unit, ax in zip(examples, axes):
    data = lkr[unit].groupby(['PG outcome', 'FG outcome']).get_group((my.HIT, my.HIT))
    for group, gdata in data.groupby(['PG port', 'FG port']):
        counts = my.analyze.time_histogram(gdata, limit=(-1,1), bin_width=bin_width)*bin_width
        
        # Here I'm going to get estimates of the spike rate during each time window
        # using a Bayesian model with a Poisson likelihood and a Gamma prior on the
        # rate. The second and third arguments set the Gamma parameters.
        # The Jeffreys' prior is (0.5, 0), which an exponential prior is (1,1)
        means, CRs = vect_rate_bayes(counts, 0.5, 0)
        
        # Now do the actual plotting
        yerrs = np.abs(CRs.T - means)/bin_width
        ax.errorbar(rate.columns.values+bin_width/2., means/bin_width, yerr=yerrs, 
                    fmt='-', capsize=0, 
                    label='{} to {}'.format(labels[group[0]], labels[group[1]]))
        ax.plot([0,0], ax.get_ylim(), '--k', alpha=0.2)
    
    ax = my.utils.pub_format(ax, yticks=3)
    ylims = ax.get_ylim()
    lq, uq = data['C in'].quantile(q=[0.25, .75])
    ax.fill_betweenx(ax.get_ylim(), lq, uq, color='steelblue', alpha=0.25, zorder=3)
    lq, uq = data['FG in'].quantile(q=[0.25, .75])
    ax.fill_betweenx(ax.get_ylim(), lq, uq, color='orange', alpha=0.25, zorder=3)
    ax.set_xlim(-1,1)
leg = axes[0].legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
                     ncol=2, mode="expand", borderaxespad=0.,
                     numpoints=1, fontsize=12)
my.utils.pub_format(axes[0], legend=leg, yticks=3, xticks=3)
axes[2].set_ylabel('Spike rate (Hz)', ha='left')
axes[-1].set_xlabel('Time from choice (s)')

fig.tight_layout(pad=0.1, h_pad=-5)
```

I'm going to make the same plot as above, but this time with a rolling window, and with a shaded region for the uncertainty.

```python
lkr.lock('C out')
examples = [53, 57, 236, 313]
fig, axes = plt.subplots(nrows=len(examples), sharex=True, figsize=(5,6))
bin_width = 0.1
labels = {my.LEFT:'Left', my.RIGHT:'Right'}
for unit, ax in zip(examples, axes):
    data = lkr[unit].groupby(['PG outcome', 'FG outcome']).get_group((my.HIT, my.HIT))
    for (group, gdata), color in zip(data.groupby(['PG port', 'FG port']), plt.rcParams['axes.color_cycle'][:4]):
        counts = my.analyze.time_histogram(gdata, limit=(-1,1), bin_width=(bin_width))*bin_width
        rates_0, CRs_0 = vect_rate_bayes(counts, 0.5, 0)
        
        # Roll the windows by shifting bins to the right by half the bin width
        counts_shift = my.analyze.time_histogram(gdata, limit=(-1+bin_width/2.,1+bin_width/2.),
                                         bin_width=(bin_width))*bin_width
        rates_1, CRs_1 = vect_rate_bayes(counts_shift, 0.5, 0)
        
        # Interleave means and CRs and bin times
        rates = np.zeros(rates_0.size + rates_1.size)
        rates[0::2], rates[1::2] = rates_0, rates_1
        CRs = np.zeros(CRs_0.size + CRs_1.size).reshape(CRs_0.shape[0]*2,2)
        CRs[0::2,:], CRs[1::2,:] = CRs_0, CRs_1
        ts = np.zeros(counts.columns.size + counts_shift.columns.size)
        ts[0::2], ts[1::2] = counts.columns.values, counts_shift.columns.values
        ts = ts + bin_width/2
        
        # Do the actual plotting
        ax.plot(ts, rates/bin_width, label='{} to {}'.format(labels[group[0]], labels[group[1]]))
        ax.fill_between(ts, CRs[:,0]/bin_width, CRs[:,1]/bin_width, alpha=0.25, color=color)
        
        ax.plot([0,0], ax.get_ylim(), '--k', alpha=0.2)
        
    ax = my.utils.pub_format(ax, yticks=3)
    ylims = ax.get_ylim()
    lq, uq = data['C in'].quantile(q=[0.25, .75])
    ax.fill_betweenx(ylims, lq, uq, color='steelblue', alpha=0.25, zorder=3)
    lq, uq = data['FG in'].quantile(q=[0.25, .75])
    ax.fill_betweenx(ylims, lq, uq, color='orange', alpha=0.25, zorder=3)
    ax.set_xlim(-1,1)
leg = axes[0].legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
                     ncol=2, mode="expand", borderaxespad=0.,
                     numpoints=1, fontsize=12)
my.utils.pub_format(axes[0], legend=leg, yticks=3, xticks=3)
axes[2].set_ylabel('Spike rate (Hz)', ha='left')
axes[-1].set_xlabel('Time from choice (s)')

fig.tight_layout(pad=0.1, h_pad=-5)
```

```python
lkr.lock('C out')
examples = [53, 255, 57, 236, 135]
fig, axes = plt.subplots(nrows=len(examples), sharex=True, figsize=(4,5))
for unit, ax in zip(examples, axes):
    data = lkr[unit].groupby(['PG outcome', 'FG outcome']).get_group((my.HIT, my.HIT))
    rates = [my.analyze.smooth(gdata, limit=(-1,1), width=0.03)
             for group, gdata in data.groupby('PG port')]
    means = np.array([each.mean().values for each in rates])
    
    # Plot standardized activity
    xs = rates[0].columns.values
    standard = sp.stats.zscore(means, axis=1).T
    lines = ax.plot(xs, standard)
    
    # Fill in significantly different activity
    ps = np.array([sp.stats.mannwhitneyu(rates[0][each], rates[1][each])[1]
                   for each in xs])
    crit_p = 0.05/float(len(ps))
    ax.fill_between(xs, standard[:,0], standard[:,1], 
                    where=(ps <= crit_p), color='k', alpha=0.3)
    ax.set_ylim(ax.get_ylim()[0] - 5, ax.get_ylim()[1]+1)
    ax.plot([0,0], ax.get_ylim(), '--k', alpha=0.5)

for ax in axes[1:-1]:
    ax = my.utils.pub_format(ax, hide_spines=['left', 'top', 'bottom', 'right'])
    ax.set_yticklabels('')

# Format top axis
leg = axes[0].legend(lines, ('Left', 'Right'), bbox_to_anchor=(0., 1.302, 1., .402), loc=8,
               ncol=2, borderaxespad=0., numpoints=1, fontsize=12)
ax = my.utils.pub_format(axes[0], legend=leg, hide_spines=['left', 'top', 'bottom', 'right'])
ax.set_yticklabels('')

# Format bottom axis
ax = my.utils.pub_format(axes[-1], hide_spines=['left','top','right'])
ax.set_yticklabels('')
ax.set_xlabel('Time from choice (s)')

fig.tight_layout(pad=0.1, h_pad=0)
```
